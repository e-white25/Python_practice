Compound Interest can be expressed with the formula $$A = P (1 + \frac{r}{n})^{nt}$$
- $A$ is amount accumulated
- $P$ is principal
- $r$ is interest rate
- $n$ is compounds per period
- $t$ is number of periods

````python
### Write a function to calculate compund interest
### Note: assume it compuntds monthly
### Note: input params for functions should be principal, time, and interest.
### Time: # of years

n = 12

def interest(principal, time, interest):
  return principal * (1 + interest/n)**(n*time)
````

````python
### call this function at 2, 4, 10 years print the values use simple formatting in all cases use 5% interest rate and a initial deposit of 1000
total = interest(1000, 2, .05)
print(total)
````

````python
#### SIMPLE EXERCISE:
## Write a function that takes a single string as input and calculates the GC content of it
## GC content is the percentage (or fraction) of a sequence that it is G or C.

def GC_content(seq):
    # Count the occurrences of 'G' and 'C' in the input sequence
    G_count = seq.count('G')
    C_count = seq.count('C')

    # Calculate the total count of G and C
    total_gc = G_count + C_count

    # Calculate the GC content as a percentage
    gc_content_percentage = (total_gc / len(seq)) * 100

    return gc_content_percentage

# Test the function with an example sequence
sequence = 'AGCTTTTCATTCTGACTGCAACGGGCAATATGTCTCTGTGTGGATTAAAAAAAGAGTGTCTGATAGCAGC'
print(GC_content(sequence))
````
